```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
x1 = np.array([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5,5.5, 6])
x0 = np.ones(x1.shape)
y  = np.array([0.17,0.28,0.39,0.47,0.43,0.60,0.74,0.79,0.88,0.84,0.96])

X = np.matrix([x0,x1]).T
Y = np.matrix([y]).T

#Regresi贸n con ecuaci贸n normal.

theta = np.linalg.inv(X.T*X)*(X.T)*Y

plt.plot(x1,y, 'bo')
plt.plot(x1,theta[0,0]+theta[1,0]*x1)
plt.title("Regresi贸n de voltaje vs corriente")
r_cuadrado = r2_score(y,theta[0,0]+theta[1,0]*x1)
plt.show()

print("Coeficiente de determinaci贸n R2: ", r_cuadrado)

```
